# Import

First, export your MyAnimeList by visting https://myanimelist.net/panel.php?go=export. 

Click the "Export My List" button. A XML file should have been downloaded. 

Change the address variable below to match the XML file's path.

Now, just run all code cells.

```
import requests
from bs4 import BeautifulSoup

import xml.etree.ElementTree as ET

# read the XML file
address = '/content/drive/My Drive/Colab Notebooks/animelist_1593694079_-_6819869.xml'
tree = ET.parse(address)
root = tree.getroot()
```

```
def url_formatter(title):
  # removal of punctuation
  # (, ), !, ?, ., ;, *, ,, ',
  title = title.replace('(', '')
  title = title.replace(')', '')
  title = title.replace('!', '')
  title = title.replace('?', '')
  title = title.replace('.', '')
  title = title.replace(';', '')
  title = title.replace('*', '')
  title = title.replace(',', '')
  title = title.replace('\'', '')


  # replace with underscore
  # :, &, space
  title = title.replace(' ', '_')
  title = title.replace('&', '_')
  title = title.replace(':', '_')

  return title
```

```
# iterate through each anime
for child in root:

  # only anime that are watching or watched
  if child[12].text == 'Completed' or child[12].text == 'Watching':


    title = url_formatter(child[1].text)
    URL = 'https://myanimelist.net/anime/' + child[0].text + '/' + title

    # get HTML file from URL
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')

    try:
      # look for the op songs
      op_comb = soup.find('div', class_= 'theme-songs js-theme-songs opnening')

      # get rid of <div class> html
      op_sep = op_comb.find_all('span', class_ = 'theme-song')

      
      
      for song in op_sep:

        # string manipulaton
        song_name = song.text

        # if numbered song
        if '#' == song_name[0]:
          # then remove the number
          index = song_name.find(':')+2
          song_name = song_name[index:]

        print(song_name)

    except:
      print(child[1].text + " does not have an OP!!!!!!!")


    try:
      # look for ed songs
      ed_comb = soup.find('div', class_= 'theme-songs js-theme-songs ending')

      # get rid of <div class> html
      ed_sep = ed_comb.find_all('span', class_ = 'theme-song')

      for song in ed_sep:

        # string manipulaton
        song_name = song.text

        # if numbered song
        if '#' == song_name[0]:
          # then remove the number
          index = song_name.find(':')+2
          song_name = song_name[index:]

        print(song_name)
    
    except:
      print(child[1].text + " does not have an ED!!!!!!!")
```
